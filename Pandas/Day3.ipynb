# Работа с пропусками и дубликатами(df.isnull().sum(), dropna(), fillna(), duplicated(), drop_duplicates())

```python
# Создадим пропуски в числовых колонках (например, 5 случайных значений в age и bmi)
df.loc[np.random.choice(df.index, 5, replace=False), 'age'] = np.nan
df.loc[np.random.choice(df.index, 5, replace=False), 'bmi'] = np.nan

# Создадим пропуски в категориальных колонках (sex, smoker)
df.loc[np.random.choice(df.index, 3, replace=False), 'sex'] = np.nan
df.loc[np.random.choice(df.index, 3, replace=False), 'smoker'] = np.nan
```
