# df.groupby('col').mean()

```python
df2 = pd.DataFrame({
    'Team': ['A', 'B', 'A', 'B', 'A'],
    'Player': ['Alice','Bob','Charlie','David','Eva'],
    'Score': [10, 20, 30, 40, 50]
})
df2
```

```python
team_mean = df2.groupby('Team')['Score'].min()
print(team_mean)
```

```python
team_mean = df2.groupby('Team')['Score'].mean()
print(team_mean)
```

```python
team_mean = df2.groupby('Team')['Score'].count()
print(team_mean)
```

```python
team_mean = df2.groupby('Team')['Score'].sum()
print(team_mean)
```
