### Block 1(apply)  take a function and apply it to each element 

```python
df['age_plus1'] = df['Age'].apply(lambda x: x + 1)
```

```python
df.head(5)
```

```python
def age_group(age):
    if age < 18: 
        return 'child'
    elif age <= 50:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(age_group)
df[['Age', 'age_group']].head(5)
```

```python
df['age_group'] = df['Age'].apply(
    lambda x: 'child' if x < 18 else ('adult' if x <= 50 else 'senior') # тернарный оператор 
)
```

```python
df
```

```python
def survival_type(row):
    if row['Survived'] == 1 and row['Fare'] > 50:
        return 'rich_survived'
    elif row['Survived'] == 1:
        return 'poor_survived'
    else :
        return 'not_survived'
df['survival_type'] = df.apply(survival_type, axis=1) ## axis=1 If you use multiple columns at the same time
df[['Survived', 'Fare', 'survival_type']].head(5)
```

```python
df['survival_type'] = df.apply(
    lambda r: 'rich_survived' if r['Survived'] == 1 and r['Fare'] > 50
    else ('poor_survived' if r['Survived'] == 1 else 'not_survived'),
    axis = 1
)
```

```python
df.head(5)
```

```python
def family_size(row):
    return row['SibSp'] + row['Parch'] + 1 

df['family_size'] = df.apply(family_size, axis=1)
df[['SibSp', 'Parch', 'family_size']].head(5)
```

```python
def avgFare_family(row):   ## an average fare for 1 person from family
    return ((row['Fare']) / (row['Parch'] + row['SibSp'] + 1))

df['avgFare_family'] = df.apply(avgFare_family, axis=1)
df[['Parch', 'SibSp', 'Fare', 'avgFare_family']].head(5)
```

```python
df.head(5)
```

```python
df['fare_level'] = df.apply(
    lambda x: 'low' if x['Fare'] < 20
    else('medium' if 20 <= x['Fare'] <= 50 else 'high'),
    axis = 1
)
```

```python
df.tail(5)
```

```python
df['survived_female'] = df.apply(
    lambda x: 1 if x['Sex'] == 'female' and x['Survived'] == 1
    else 0,
    axis=1
)
df[['Survived', 'Sex', 'survived_female']]
```

```python
df['survived_female'] = np.where(
     (df['Sex'] == 'female') & (df['Survived'] == 1), 1, 0
)
df[['Survived', 'Sex', 'survived_female']]
```

```python
df['priority'] = np.where(
    (df['Sex'] == 'female') | (df['Age'] < 16), 1, 0
)
df[['Sex', 'Age', 'priority']].head(10)
```

```python
df1 = pd.read_csv('/Users/danialmendigali/jupyter-datascience/Pandas/data set /insurance.csv')
```

```python
def charges_level(row):
    if row['charges'] < 3000:
        return 'low'
    elif row['charges'] > 3000 and row['charges'] < 10000:
        return 'medium'
    else:
        return 'high'

df1['charges_level'] = df1.apply(charges_level, axis=1)
df1.head(5)
```

```python
df1['risk_level'] = df1.apply(
    lambda x: 'low'
    if x['smoker'] == 'no'
    else('medium' if 17 < x['bmi'] < 30 else 'high'),
    axis=1 
)
```

```python
def risk_level(row):
    if row['smoker'] == 'yes' and df1['bmi'].min() < row['bmi'] < 30:
        return 'medium'
    elif row['smoker'] == 'no':
        return 'low'
    else:
        return 'high'

df1['risk_level'] = df1.apply(risk_level, axis=1)
df1[['bmi', 'smoker', 'risk_level']].loc[
    (df1['smoker'] == 'yes')
].head(5)
```
